### Dicionario de dados

#### Dataset clinics

- CLINIC_ID

    Identificador da clinica

- CLINIC_CREATED_AT

    Data e horário de cadastro em que a clínica foi cadastrada no sistema

- BUSINESS_SEGMENTATION

    Segmentação do negócio ['Independente', 'Rede Homologada', 'Independente - Legado']

- IS_CHAIN_CLINIC

    É uma rede de clinica? [False,  True]

- FIRST_USER_HUBSPOT_SOURCE

    Fonte inicial pela qual o usuário descobriu ou foi adquirido pela clínica (por exemplo, publicidade, referência, etc.). ['Inbound', 'Redes', 'no_user', 'Outbound', 'Eventos']

- INTEREST_REASONS

    Razões de interesse expressas pela clínica.

    - Motivos de nulos: Aparentemente os valores nulos se encaixam na categoria de 'Desconhecido'

- INTEREST_CATEGORY_SIGNUP

    Categoria de cadastro desejada. ['BNPL e SaaS', 'Apenas BNPL', 'Apenas SaaS', 'Desconhecido']
    
    - BNPL = Buy Now Pay Later;

    - SaaS = Software as a Service

- HAS_INTEREST_BNPL

    Tem interesse em BNPL? [True, False]

- HAS_INTEREST_BNPL_ONLY

    Tem interesse em somente BNPL? [True, False]

- HAS_INTEREST_SAAS

    Tem interesse em SaaS? [True, False]

- TRIAL_DURATION

    Duração do periodo trial em dias

- TRIAL_START_DATE

    Data de começo do periodo trial [YYYY, MM, DD]


- TRIAL_END_DATE   

    Data de fim do periodo trial [YYYY, MM, DD]

- TRIAL_START_HOUR

    Hora do começo do periodo trial

- TRIAL_START_HOUR_CATEGORY

    Categoria da hora ['After Hours', 'Business Hours']

- TRIAL_START_DAY_OF_WEEK

    Dia da semana que começou o trial

- TRIAL_START_DAY_OF_WEEK_CATEGORY

    Categoria do dia da semana ['Weekday', 'Weekend']

- HAS_USER_CREATED

    Indica se um usuário foi criado dentro do sistema da clínica. [True, False]

- MARKETING_ATTRIBUITION

    Atribuição ao marketing ['Paid - meta', 'Paid - google', 'Other', 'Site - pediu ajuda', 'Organic - search', 'Indicação - interna', 'Organic - referral', 'Organic - social']

- MARKETING_ATTRIBUITION_CHANNEL_GROUP

    Grupo de marketing ['Paid - meta', 'Paid - google', 'Other', 'Other - known', 'Organic']

- MARKETING_ATTRIBUITION_CAMPAIGN

    Campanha de marketing

    - Motivos de nulos: Não houve campanha de marketing


- MARKETING_ATTRIBUITION_CAMPAIGN_PRODUCT

    Produto da campanha de marketing ['BNPL', 'SaaS', nan, 'Branding']

    - Motivos de nulos: Não houve campanha de marketing

- MARKETING_ATTRIBUITION_AD_GROUP

    Grupo de anúncios da campanha de marketing que levou à aquisição do usuário ['interesses-validados', nan, 'lal-all-1', 'base-fria-55k-1', 'capim', 'kw-software-odontologico', 'kw-marca', 'termos-software', 'conselhos-lal']

    - Motivos de nulos: Não houve campanha de marketing

- IS_PAID_MEDIUM

    Indica se o meio pelo qual o usuário foi adquirido era pago. [True, False]

- HAS_ASKED_FOR_ACCREDITATION

    Pediu por credenciamento? [False,  True]

- IS_ACCREDITATION_APPROVED

    Credenciamento aprovado? [False,  True]

- IS_ACCREDITATION_REPROVED

    Credenciamento foi reprovado?

- CLINIC_ACCREDITATION_STATUS

    Status de credenciamento da clinica [nan, 'Incompleta', 'Concluído', 'Reprovado', 'Concluído - faltam dados bancários', 'Com pendência (fase 2)']

- REQUEST_FINISHED_AT

    Data de finalização do pedido

- ACCREDITATION_REQUESTED_AT

    Data do pedido de credenciamento

- ACCREDITATION_APPROVED_AT

    Data de aceitação do credenciamento

- ACCREDITATION_REJECTED_AT

    Data de rejeição do credenciamento

#### Dataset subscriptions

- CLINIC_ID

    Identificador da clinica

- STRIPE_SUBSCRIPTION_ID

    Identificador único da assinatura dentro do sistema de pagamento Stripe.

- SUBSCRIPTION_START_DATE

    Inicio da assinatura

- SUBSCRIPTION_END_DATE
    
    Termino da assinatura

- SUBSCRIPTION_CURRENT_PERIOD_STARTED_AT

    Data de início do período atual de pagamento dentro da assinatura.

- SUBSCRIPTION_CURRENT_PERIOD_ENDS_AT

    Data de término do período atual de pagamento dentro da assinatura.

- MOST_RECENT_INVOICE_CREATED_AT

    Data de criação da fatura mais recente associada à assinatura.

- LAST_PAYMENT_AT

    Data do ultimo pagamento realizado.

- CHECKOUT_STATUS

    Status do processo de pagamento/checkout.

- SUBSCRIPTION_STATUS

    Status da assinatura.

- BILLING_INTERVAL_IN_MONTHS

    Intervalo das contas em meses [ 1., 12.,  6., nan]

- FIRST_PAYMENT_METHOD_TYPE

    Metodo do primeiro pagamento

- FIRST_CARD_BRAND

    Bandeira do cartão ['mastercard', 'visa', nan]

- FIRST_PAYMENT_AMOUNT

    Valor do primeiro pagamento

- FIRST_PAYMENT_AMOUNT_OFF

    Desconto do valor do primeiro pagamento

- FIRST_PAYMENT_PROMOTION_CODE

    Cupom de codigo para o primeiro pagamento

- FIRST_PAYMENT_COUPON_ID

    Identificador do cupom

- FIRST_PAYMENT_AMOUNT_OFF_2 

    Outro desconto do valor do primeiro pagamento


#### Dataset activity

- CLINIC_ID

    Identificador da clinica

- ACTIVITY_AT

    Data e hora que a atividade foi feita

- ACTIVITY_TYPE

    Tipo da atividade

- FEATURE

    O aspecto, caracteristica da atividade

- MODULE

    Módulo do sistema ou parte do produto em que a atividade ocorreu.

- ANALYTICS_DOMAIN

    Domínio analítico ['patient', 'scheduling', 'setup', 'finance', 'bnpl']

- IS_TRANSACTION

    É transação? [False,  True]

- IS_BNPL_ACTIVITY

    É uma atividade BNPL? [False,  True]

- IS_VALID_ACTIVITY

    É uma atividade valida? [True, False]

- IS_DELETION_ACTIVITY

    É uma atividade de eliminação? [True, False]

- IS_MANAGEMENT_ACTIVITY

    É uma atividade de gerenciamento? [True, False]

- IS_FINANCE_ACTIVITY

    É uma atividade de finança? [True, False]
